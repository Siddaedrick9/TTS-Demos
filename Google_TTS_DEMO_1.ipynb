```python
# Install googletrans (ensure you're using version 4.0.0-rc1)
#!pip install googletrans==4.0.0-rc1

from googletrans import Translator

# Initialize the Translator
translator = Translator()

# Text to translate
text = "Hello, how are you?"

# Translate text from English ('en') to Hindi ('hi')
translated = translator.translate(text, src='en', dest='hi')

print("Original:", text)
print("Translated:", translated.text)
```

```python


from gtts import gTTS
import tempfile
from IPython.display import Audio, display

def text_to_speech(text, lang='en'):
    """
    Converts input text to speech, saves it to a temporary MP3 file, and returns the file path.

    Parameters:
        text (str): The text to convert.
        lang (str): Language for the TTS (default is English, 'en').

    Returns:
        str: The path to the temporary MP3 file.
    """
    # Create a gTTS object
    tts = gTTS(text=text, lang=lang)
    # Create a temporary file to save the audio
    temp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
    tts.save(temp_file.name)
    return temp_file.name

# Get user input text (In Google Colab, you can use an input widget or assign the text directly)
user_text = input("Enter text to convert to speech: ")

# Convert the text to speech and get the audio file path
audio_file = text_to_speech(user_text)

# Display and play the audio file in the notebook
print("Playing audio...")
display(Audio(audio_file, autoplay=True))
```
